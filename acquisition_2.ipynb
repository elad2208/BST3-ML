```python
import requests
url='https://www.basketball-reference.com/leagues/NBA_1990_per_poss.html'
html = requests.get(url)

from bs4 import BeautifulSoup
import pandas as pd
soup = BeautifulSoup(html.content, "html.parser")

```

```python
players=list()
team_s=list()
game_s=list()
gs_s=list()
mp_s=list()
fg_s=list()
fga_s=list()
fgp_s=list()
trb_s=list()
stl_s=list()
pf_s=list()
ortg_s=list()
drtg_s=list()

table = soup.find('table')
table_rows = table.find_all('tr') 
```

```python

for tr in table_rows:
    td = tr('td')
    row = [tr.text.strip() for tr in td if tr.text.strip()]
    if row:
        players.append(row[0])
        team_s.append(row[1])
        game_s.append(row[2])
        gs_s.append(row[3])
        mp_s.append(row[4])
        fg_s.append(row[5])
        fga_s.append(row[6])
        fgp_s.append(row[7])
        trb_s.append(row[8])
        stl_s.append(row[9])
        pf_s.append(row[10])
        ortg_s.append(row[11])
        drtg_s.append(row[12])
```

```python

df = pd.DataFrame({"player":players,"Tm":team_s,"G":game_s,"GS":gs_s,"MP":mp_s,"FG":fg_s,"FGA":fga_s,"FG%":fgp_s,"TRB":trb_s,"STL":stl_s,"PF":pf_s,"ORTG":ortg_s,"DRTG":drtg_s})
df
        
```
